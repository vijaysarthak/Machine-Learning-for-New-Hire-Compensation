```python
# Import the pandas library for data manipulation and analysis
# Import the numpy library for numerical operations and array processing
# Import the seaborn library for data visualization

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('D:/Internshala/Python with ML/Project/Data/ML case Study.csv')
college = pd.read_csv('D:/Internshala/Python with ML/Project/Data/Colleges.csv')
cities = pd.read_csv('D:/Internshala/Python with ML/Project/Data/cities.csv')
```

```python
df.head()
```

```python
college.head()
```

```python
cities.head()
```

```python
# Extract data from the "Tier 1," "Tier 2," and "Tier 3" columns of the 'college' DataFrame and store them in separate lists 'Tier1,' 'Tier2,' and 'Tier3' for further analysis.

Tier1 = college["Tier 1"].tolist()
Tier2 = college["Tier 2"].tolist()
Tier3 = college["Tier 3"].tolist()
```

```python
Tier1
```

```python
Tier2
```

```python
# Assign tier values to colleges in the DataFrame based on their tier classification
# - If a college is in 'Tier1', set the value to 3
# - If a college is in 'Tier2', set the value to 2
# - If a college is in 'Tier3', set the value to 1
# Tier1 college get value of 3 and tier 3 of 1 because tier1 college has higher weightage then 2 and 3.

for item in df.College:
    if item in Tier1:
        df["College"] = df["College"].replace(item, 3)
    elif item in Tier2:
        df["College"] = df["College"].replace(item, 2)
    elif item in Tier3:
        df["College"] = df["College"].replace(item, 1)
```

```python
df.head()
```

```python
# Extracting lists of metropolitan and non-metropolitan cities from the 'cities'

metro = cities['Metrio City'].tolist()
non_metro_cities = cities['non-metro cities'].tolist()
```

```python
metro
```

```python
non_metro_cities
```

```python
# Assigning value as 1 if city is merto and 0 if non metro

for item in df.City:
    if item in metro:
        df['City'] = df['City'].replace(item,1)
    elif item in non_metro_cities:
        df['City'] = df['City'].replace(item,0)
```

```python
df.head()
```

```python
# Converting Categorical column into numerical by using dummy.

df = pd.get_dummies(df, drop_first=True, dtype = int)
```

## Applying Machine Learning models
Here I will applying all possible algorithm without any scaling to check performance of mode

```python
# Import necessary libraries

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
# Spliting the datas in X-y

X = df.loc[:, df.columns != 'CTC']
y = df['CTC']
```

```python
# Split Data into train and test with test_size = 0.2(80% data into train and 20% to test)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 36)
```

```python
y_test
```

### Linear Regression Model

```python
# Creating a LinearRegression model
lin_reg = LinearRegression()

# Fiting the model
lin_reg.fit(X_train, y_train)

# Predicting the test data
y_train_pred = lin_reg.predict(X_train)

y_test_pred = lin_reg.predict(X_test)
```

```python
r2_score(y_train, y_train_pred)
```

```python
# Calculate r2_score
r2_score(y_test, y_test_pred)
```

```python
# Calculate Mean Absolute Error (MAE)
mean_absolute_error(y_test, y_test_pred)
```

```python
# Calculate Mean Squared Error (MSE)
mean_squared_error(y_test, y_test_pred)
```

```python
# Print the coefficients and intercept of the linear regression model
lin_reg.coef_
```

```python
# Print the intercept of the linear regression model
lin_reg.intercept_
```
